```python
import json
import os
from tqdm import tqdm_notebook
```

```python
c = 0
for root, dirs, files in os.walk('thesis'):

    for file in tqdm_notebook(files):
        
        if file.endswith('.ipynb'):
            c = c + 1

            open_path = os.path.join(root, file)
#             print(file)
            print(open_path)
            with open(open_path) as data:
                ff = json.load(data)
                
                ff['cells'].insert(0, {'cell_type': 'code',
 'execution_count': None,
 'metadata': {},
 'outputs': [],
 'source': ['#######################################################################################\n',
  '# Author: Srijan Verma                                                              #\n',
  '# School of Pharmacy                                                                #\n',
  '# Sirimulla Research Group [http://www.sirimullaresearchgroup.com/]                 #\n',
  '# The University of Texas at El Paso, TX, USA                                       #\n',
  '# Last modified: 19/12/2019                                                         #\n',
  '# Copyright (c) 2019 Srijan Verma and Sirimulla Research Group, under MIT license   #\n',
  '#######################################################################################']})
                
                save_path = os.path.join(root, file)
                print(save_path)
                with open(save_path, 'w') as fp:
                    json.dump(ff, fp, indent=4)
                    print(c)
                
#             print(file)
#             print(root)
```
